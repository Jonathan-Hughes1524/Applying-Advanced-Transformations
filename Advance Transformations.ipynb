```python
#standard imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#importing the os and json modules
import os, json
```

```python
hero = pd.read_csv('Data/superhero_info - superhero_info.csv')
hero.head()
```

```python
power = pd.read_csv('Data/superhero_powers - superhero_powers.csv')
power.head()
```

```python
#split hero/publisher
hero['Hero|Publisher'].str.split('|',expand=True)
```

```python
#add split columns to the hero
hero[['Hero','Publisher']]=hero['Hero|Publisher'].str.split('|',expand=True)
hero.head(2)
```

```python
#drop original column
hero = hero.drop(columns=['Hero|Publisher'])
hero.head(2)
```

```python
#seperate measurments
hero.loc[0,'Measurements']
```

```python
#fix to use json
hero['Measurements']= hero['Measurements'].str.replace("'",'"')
```

```python
hero['Measurements']
```

```python
#json.loads()
hero['Measurements']= hero['Measurements'].apply(json.loads)
hero.loc[0,'Measurements']
```

```python
#make measurements into seperate columns
height_weight = hero['Measurements'].apply(pd.Series)
height_weight
```

```python
#add columns to original
hero = pd.concat((hero,height_weight),axis=1)
hero.head(2)
```

```python
#drop measurements column
hero = hero.drop(columns='Measurements')
hero.head(2)
```

```python
# remove cm from height
hero['Height']= hero['Height'].str.replace('cm','')
hero['Height'].head(2)
```

```python
#remove kg from weight
hero['Weight']= hero['Weight'].str.replace('kg','')
hero['Weight'].head(2)
```

```python
#convert height and weight to numerica
hero['Height']= pd.to_numeric(hero['Height'])
hero['Weight']= pd.to_numeric(hero['Weight'])
```

```python
hero.info()
```

```python
#One hot encode powers
power['Powers'].head(2)
```

```python
power.loc[0,'Powers']
```

```python
#split string
power['powers_split']=power['Powers'].str.split(',',expand=False)
power['powers_split']
```

```python
#df explode to create columns
exploded = power.explode('powers_split')
```

```python
exploded
```

```python
#save unique values from exploded
cols_to_make = exploded['powers_split'].dropna().unique()
cols_to_make
```

```python
#Loop for columns
for col in cols_to_make:
    power[col]=power['Powers'].str.contains(col)
power.head()
```

```python
#merge to dataframes
merged = pd.merge(hero,power, left_on='Hero',right_on='hero_names')
merged.head()
```

Question 1. Compare the average weight of super powers who have Super Speed to those who don't.

```python
merged.groupby('Super Speed')['Weight'].mean()
```

```python
sns.barplot(data=merged, x ='Super Speed', y='Weight');
```

Question 2. What is the average height of heros for each publisher?

```python
merged.groupby('Publisher')['Height'].mean()
```

```python
ax = sns.barplot(data=merged, x ='Publisher',y='Height')
ax.set_xticklabels(ax.get_xticklabels(),rotation=90);
```

1. On average super heros that have the super speed ability weigh more than those who don't.

2. On average Image Comics have the super heros with the tallest average height.
